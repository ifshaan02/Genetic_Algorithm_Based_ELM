```python
#GA-ELM
from sklearn import metrics
import random
from sklearn.model_selection import KFold
```

```python
# Initialize ELM (Randomly initialize weights and biases) 
def initialize_elm(n_features, n_hidden_neurons): 
    W = np.random.randn(n_features, n_hidden_neurons) 
    b = np.random.randn(n_hidden_neurons) 
    return W, b 

# Activation function for ELM (Sigmoid, Tanh, ReLU)
def activation_function(X, W, b, activation="sigmoid"): 
    net_input = np.dot(X, W) + b 
    if activation == "sigmoid": 
        return 1 / (1 + np.exp(-net_input))  # Sigmoid activation 
    elif activation == "tanh": 
        return np.tanh(net_input)  # Tanh activation 
    elif activation == "relu": 
        return np.maximum(0, net_input)  # ReLU activation 
    else: 
        raise ValueError(f"Unknown activation function: {activation}") 

# Train the ELM model 
def train_elm(X_train, y_train, n_hidden_neurons, activation="sigmoid"): 
    n_features = X_train.shape[1] 
    W, b = initialize_elm(n_features, n_hidden_neurons) 
    H = activation_function(X_train, W, b, activation) 
    beta = np.linalg.pinv(H).dot(y_train) 
    return W, b, beta 

# Predict using the trained ELM model 
def predict_elm(X_test, W, b, beta, activation="sigmoid"): 
    H = activation_function(X_test, W, b, activation) 
    return np.dot(H, beta)

# Evaluate the fitness (RMSE or other metrics) of the population using k-fold cross-validation 
def evaluate_population(population, X, y, k=5, evaluation_metric="rmse"): 
    fitness = [] 
    kf = KFold(n_splits=k, shuffle=True, random_state=42) 
    
    # Ensure y is a numpy array 
    y = np.array(y) 

    for n_hidden_neurons in population: 
        fold_scores = [] 
        for train_index, val_index in kf.split(X): 
            X_train, X_val = X[train_index], X[val_index] 
            y_train, y_val = y[train_index], y[val_index] 
             
            # Train the ELM model 
            W, b, beta = train_elm(X_train, y_train, n_hidden_neurons) 
             
            # Make predictions 
            prediction = predict_elm(X_val, W, b, beta) 
             
            # Calculate the specified evaluation metric
            if evaluation_metric == "rmse":
                score = np.sqrt(metrics.mean_squared_error(y_val, prediction))
            elif evaluation_metric == "mae":
                score = metrics.mean_absolute_error(y_val, prediction)
            elif evaluation_metric == "r2":
                score = metrics.r2_score(y_val, prediction)
            else:
                raise ValueError(f"Unsupported evaluation metric: {evaluation_metric}")

            fold_scores.append(score)
        
        # Average score over all folds for this individual
        avg_score = np.mean(fold_scores)
        fitness.append(avg_score) 
    
    return fitness 

# Create the initial population (Random number of hidden neurons between 5 and 200) 
def create_population(pop_size, min_neurons=5, max_neurons=200): 
    return [random.randint(min_neurons, max_neurons) for _ in range(pop_size)]  # Each individual represents neurons 

# Select the best parents (individuals with lowest RMSE)
def select_parents(population, fitness, num_parents): 
    selected_parents = [population[i] for i in np.argsort(fitness)[:num_parents]] 
    return selected_parents 

# Crossover operation (one-point crossover)
def crossover(parents, crossover_prob): 
    if random.random() < crossover_prob: 
        offspring = [parents[0], parents[1]]  # Swap values between parents to create offspring 
        return offspring 
    return parents

# Mutation operation (alter the number of neurons with a given probability)
def mutate(population, mutation_prob, min_neurons=5, max_neurons=200): 
    for i in range(len(population)): 
        if random.random() < mutation_prob: 
            population[i] = random.randint(min_neurons, max_neurons)  # Change the number of hidden neurons
    return population 

# Genetic Algorithm Parameters (dynamic nature)
def run_genetic_algorithm(X_train, y_train, X_test, y_test, pop_size=10, num_generations=10, 
                          num_parents=4, crossover_prob=0.7, mutation_prob=0.2, 
                          evaluation_metric="rmse", min_neurons=5, max_neurons=200):
    # Create the initial population 
    population = create_population(pop_size, min_neurons, max_neurons) 
    
    # Run the Genetic Algorithm
    for generation in range(num_generations): 
        print(f"Generation {generation+1}") 
        
        # Evaluate the current population with k-fold cross-validation
        fitness = evaluate_population(population, X_train, y_train, k=5, evaluation_metric=evaluation_metric)
        
        # Select the best parents
        parents = select_parents(population, fitness, num_parents)
        
        # Generate the next generation using crossover
        offspring = []
        while len(offspring) < pop_size - num_parents: 
            parent_pair = random.sample(parents, 2) 
            offspring.extend(crossover(parent_pair, crossover_prob))
        
        # Apply mutation
        offspring = mutate(offspring, mutation_prob, min_neurons, max_neurons)
        
        # Create the new population by combining parents and offspring
        population = parents + offspring
        
        # Print the best RMSE for each generation
        best_fitness = min(fitness)
        print(f"Best {evaluation_metric.upper()} in Generation {generation+1}: {best_fitness}")
    
    # Get the best individual from the final population
    best_individual = population[np.argmin(fitness)] 
    print(f"\nBest number of hidden neurons: {best_individual}") 
    
    # Train the final ELM model with the best parameters
    W, b = initialize_elm(X_train.shape[1], best_individual) 
    H = activation_function(X_train, W, b) 
    beta = np.linalg.pinv(H).dot(y_train)
    
    # Make predictions on the test set
    predictions = predict_elm(X_test, W, b, beta)
    
    # Calculate RMSE (or other specified metric) on the test set
    if evaluation_metric == "rmse":
        rmse_test = np.sqrt(metrics.mean_squared_error(y_test, predictions))
        print(f"Final Test {evaluation_metric.upper()}: {rmse_test}")
    elif evaluation_metric == "mae":
        mae_test = metrics.mean_absolute_error(y_test, predictions)
        print(f"Final Test {evaluation_metric.upper()}: {mae_test}")
    elif evaluation_metric == "r2":
        r2_test = metrics.r2_score(y_test, predictions)
        print(f"Final Test {evaluation_metric.upper()}: {r2_test}")
```

```python
import tkinter as tk
from tkinter import messagebox, ttk
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("AirQualityUCI.csv", sep=',', engine='python', on_bad_lines='skip')
df.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1, inplace=True)

df.columns = ['DATE','TIME','CO_GT','PT08_S1_CO','NMHC_GT','C6H6_GT','PT08_S2_NMHC',
              'NOX_GT','PT08_S3_NOX','NO2_GT','PT08_S4_NO2','PT08_S5_O3','T','RH','AH']
df.dropna(inplace=True)

X = df.drop(['DATE', 'TIME'], axis=1)
y = df['RH']

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train models
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.4, random_state=42)

linear_model = LinearRegression().fit(X_train, y_train)
logistic_model = LogisticRegression(max_iter=2000, solver='saga').fit(X_train, (y_train > y_train.mean()).astype(int))
svm_model = SVR().fit(X_train, y_train)

# GA-ELM Logic
def ga_elm_predict(X):
    hidden_nodes = 10
    input_weights = np.random.randn(hidden_nodes, X.shape[1])
    biases = np.random.randn(hidden_nodes)
    H = np.tanh(np.dot(X, input_weights.T) + biases)
    output_weights = np.linalg.pinv(H).dot(y_train[:H.shape[0]])
    return H.dot(output_weights)

# Function to calculate RMSE and show graph
def calculate_rmse():
    try:
        # Collect user inputs
        feature_values = pd.DataFrame([float(entry.get()) for entry in feature_entries],
                                      index=features).T

        # Standardize the user input data
        feature_values_scaled = scaler.transform(feature_values)

        # Predictions
        y_pred_linear = linear_model.predict(feature_values_scaled)
        y_pred_logistic = logistic_model.predict(feature_values_scaled)
        y_pred_svm = svm_model.predict(feature_values_scaled)
        y_pred_ga_elm = ga_elm_predict(feature_values_scaled)

        # Calculate RMSE
        rmse_linear = np.sqrt(mean_squared_error([y_pred_linear]*len(y_test), y_test))
        rmse_logistic = np.sqrt(mean_squared_error([y_pred_logistic]*len(y_test), y_test))
        rmse_svm = np.sqrt(mean_squared_error([y_pred_svm]*len(y_test), y_test))
        rmse_ga_elm = np.sqrt(mean_squared_error([y_pred_ga_elm]*len(y_test), y_test))

        # Display results in the GUI
        result_text.set("Comparing RMSE values of our algorithm (GA-ELM) with other Machine Learning Algorithms:\n"
                f"Linear Regression: {rmse_linear:.4f}\n"
                f"Logistic Regression: {rmse_logistic:.4f}\n"
                f"Support Vector Machine: {rmse_svm:.4f}\n"
                f"GA-ELM: {rmse_ga_elm:.4f}")

        # Plot the RMSE values
        algorithms = ['Linear', 'Logistic', 'SVR', 'GA-ELM']
        rmse_values = [rmse_linear, rmse_logistic, rmse_svm, rmse_ga_elm]

        plt.figure(figsize=(8, 5))
        bars = plt.bar(algorithms, rmse_values, color=['#3498DB', '#E67E22', '#9B59B6', '#2ECC71'])
        plt.ylabel("RMSE")
        plt.title("Comparison of RMSE values of our algorithm (GA-ELM) with other Machine Learning Algorithms")
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, f'{yval:.2f}', ha='center', va='bottom', fontsize=10)
        plt.tight_layout()
        plt.show()

    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numeric values for all features.")

# Initialize GUI
root = tk.Tk()
root.title("AIR QUALITY INDEX")
root.geometry("500x600")
root.configure(bg="#2C3E50")

# Make window pop up on top
root.lift()
root.attributes('-topmost', True)
root.after_idle(root.attributes, '-topmost', False)

# Header Section
title_label = tk.Label(root, text="AIR QUALITY INDEX CALCULATOR", font=("Arial", 18, "bold"), bg="#2980B9", fg="white", padx=10, pady=10)
title_label.pack(fill="x")

# Feature labels and input fields
features = ['CO_GT','PT08_S1_CO','NMHC_GT','C6H6_GT','PT08_S2_NMHC','NOX_GT','PT08_S3_NOX',
            'NO2_GT','PT08_S4_NO2','PT08_S5_O3','T','RH','AH']

frame = tk.Frame(root, bg="#2C3E50")
frame.pack(pady=10)

feature_entries = []
for i, feature in enumerate(features):
    label = tk.Label(frame, text=f"{feature}:", font=("Arial", 10), bg="#2C3E50", fg="white")
    label.grid(row=i, column=0, padx=10, pady=5, sticky="e")
    entry = ttk.Entry(frame, width=10)
    entry.grid(row=i, column=1, padx=10, pady=5)
    feature_entries.append(entry)

# Button to calculate RMSE
calculate_button = tk.Button(root, text="Calculate AQI", command=calculate_rmse, font=("Arial", 12, "bold"), bg="#27AE60", fg="white", padx=10, pady=5)
calculate_button.pack(pady=15)

# Results display
result_text = tk.StringVar()
result_label = tk.Label(root, textvariable=result_text, font=("Arial", 12), bg="#2C3E50", fg="white", justify="left")
result_label.pack(pady=10)

root.mainloop()
```
